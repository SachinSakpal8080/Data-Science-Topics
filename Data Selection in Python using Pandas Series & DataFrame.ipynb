### Introduction:

In python, Numpy & Pandas are used to manupulate data.   
Indexing operator [] & attribute operator '.' provides quick & easy access to the data acorss wide range.   
The index is like an unique address to select the database.   
Rows & columns both have indexes.   

#### Different methods to select data:
##### Object selection:
Object selection has had several user-requested additions to support more explicit location-based indexing.   
Pandas support three types of multi-axis indexing for the selection of the data.

###### Method 1: .loc
It is a primarily label based, but may also be used with a boolean array.   
We will creat a Data frame with the help of pandas and Numpy.    
In the dataframe, we are generating random numbers with the help of random functions.   
Here the index is given with label names of small alphabet and column names given with capital alphabets.   
The index contains six alphabet means we want rows and three columns, also mentioned in the ‘randn’ function.

```python
# import numpy & pandas library
import numpy as np
import pandas as pd
import random # to generate the random numbers.

import warnings
warnings.filterwarnings('ignore') # Ignore warnings

data = pd.DataFrame(np.random.randn(6, 3), index = ['a','b','c','d','e','f'], columns = ['A', 'B', 'C'])
data
```

```python
# How to select the specific rows:
data.loc[:'d',:]
# Selected first 4 rows
```

```python
# How to check the values is positive or negative in a particular row.
data.loc['a']>0
# It will return True where condition satisfies.
# It will return False where condition doesn't satisfies.
```

```python
# How to access access the only one column.
data.loc[:,'A']
# It will return the values available in the column 'A'
```

```python
# Values from the column 'B'
data.loc[:,'B']
```

```python
# Values from the column 'C'
data.loc[:,'C']
```

```python
# How to select the multiple columns using .loc.
data.loc[:,['A', 'B']]
```

###### Method 2: .iloc
It is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.   
Pandas provide various methods to get purely integer based indexing.

```python
DATA = pd.DataFrame(np.random.randn(10, 5),columns = ['A', 'B', 'C', 'D', 'E'])
DATA
```

```python
# How to select all first 8 rows for all columns.
DATA.iloc[:8]
```

```python
# Select first 4 rows for all the columns:
DATA.iloc[:4]
```

```python
# Select row 2,3 & column B to D
DATA.iloc[1:3,1:-1]
```

###### Method 3: .ix
It is used for both labels and integer-based.  
Besides pure label based and integer-based, Pandas provides a hybrid method for selections and subsetting the object using the .ix() operator.

```python
Data1 = pd.DataFrame(np.random.randn(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
Data1
```

```python
Data1.ix[1:4,['A','B']]
# index includes both values mentioned in the range
```

###### Method 4: The query() Method
DataFrame objects have a query() method that allows selection using an expression.     
You can get the value of the frame where column B has values between the values of columns A and B.

```python
Data2 = pd.DataFrame(np.random.rand(10, 5), columns=list('ABCDE'))
Data2
```

```python
# Select data where value of B is more than A & less than c
Data2.query('(A<B) & (B<C)')
```

#### Duplicate Data:
If you want to identify and remove duplicate rows in a Data Frame, two methods will help: duplicated and drop_duplicates.

###### duplicated: 
returns a boolean vector whose length is the number of rows, and which indicates whether a row is duplicated.
###### drop_duplicates: 
removes duplicate rows.

```python
Data3 = pd.DataFrame({'A': ['one', 'one', 'two', 'two', 'two', 'three', 'three', 'three', 'three'],
                    'B': ['x', 'y', 'x', 'y', 'x', 'x', 'y', 'x', 'y'],
                    'C': np.random.randn(9)})
Data3
```

```python
# Check if there are any duplicate values in the column.
Data3.duplicated('A')
# If it's duplicate then return True else False
```

```python
Data3.drop_duplicates('A')
```
